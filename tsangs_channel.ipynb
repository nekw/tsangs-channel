```python
import yfinance as yf
import pandas as pd
#ticker = "ETH-USD\\
#"
ticker = "BTC-USD"
ticker = "BTC-ETH"
ticker = "^HSI"
ticker = "^SPX"

tickers = [ticker]
df =yf.download(tickers,  start = "2012-02-01" , end = None)
print(df)
```

```python

import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# plott the points (x, y)
x = df.index
df['log'] = df['Close'].apply(np.log10)
y = df['log']
plt.plot(x, y)

# plot channel
df['i'] = range(0, len(df))
slope, intercept, r_value, p_value, std_err = stats.linregress(df['i'], y)
for i in [-2,-1,0,1,2]:
  l = intercept + slope * df['i'] + np.std(df['log'])*i
  plt.plot(l,'k-')

# labeling
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title(f'{ticker} Tsang''s Channel')
plt.show()
```
